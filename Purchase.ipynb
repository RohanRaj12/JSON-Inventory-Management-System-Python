```python
import json

fd = open("Inventory.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)

i=1
total=0
item=0
loop=1

print("\n###########**********Inventory Management System**********############\n")
while(loop==1):
    
    print("\nEnter your Choice Please\n")
    print("1. To Buy Item")
    print("2. To View Current Available Stock")
    print("3. To View Sales Made")
    print("4. To Exit")
    ch=int(input("\nEnter Your Choice: "))
    
    if(ch==1):
    
        prodid = str(input("Enter product id: "))
        if (prodid in record.keys()):
            quan=int(input("Enter the quantity: "))
            left = record[prodid]['quantity']-quan

            if (left<=0):
                print("Item OUT OF STOCK!! Please Try Later.")
            else:
                print("\n******************************************************\n")
                print("You purchased ",quan," unit of ",record[prodid]['Name'],record[prodid]['type'])
                print("Discount is: ",record[prodid]['discount'],"%")
                print("MRP is: ",record[prodid]['price'])

                name = record[prodid]['Name'] 
                types = record[prodid]['type']
                price = record[prodid]['price']
                disc = record[prodid]['discount']

                record[prodid] = {'Name': name, 'type': types , 'price': price, 'quantity': left, 'discount': disc}

                js = json.dumps(record)
                fd = open("Inventory.json",'w')
                fd.write(js)
                fd.close()

                dprice=price-(price*(disc/100))
                print("Discounted price: ",dprice)
                tbill=dprice*quan
                print("\nTotal Bill : ",tbill)

                ad = open("sales.json",'r')
                rr = ad.read()
                ad.close()

                sales = json.loads(rr)

                if i in sales.keys():
                        i=int(max(sales.keys()))+1
                        sales[i]={'Name': name, 'type': types, 'price': price, 'quantity': quan, 'discount': disc, 'totalbill': tbill}
                else:
                        sales[i]={'Name': name, 'type': types, 'price': price, 'quantity': quan, 'discount': disc, 'totalbill': tbill}
                        ab = json.dumps(sales)
                        rd = open("sales.json",'w')
                        rd.write(ab)
                        rd.close()
         
        else:
            print("Item not there")
            
    elif(ch==2):
        fd = open("Inventory.json",'r')
        r = fd.read()
        fd.close()
        record = json.loads(r)
        print("\nID\t\tNAME\t\tTYPE\t\tPRICE\t\tQUANTITY\tDISCOUNT\n")
        for i in record:
            print(i,"      ",record[i]['Name'],"\t\t",record[i]['type'],"\t",record[i]['price'],"\t\t",record[i]['quantity'],"\t\t",record[i]['discount'])
            
    elif(ch==3):
        bd = open("sales.json",'r')
        b = bd.read()
        bd.close()
        cd = json.loads(b)
        print("\nItems sold are\n")
        for i in cd:            
            print(cd[i]['quantity']," unit of ",cd[i]['Name'],"of Amount",cd[i]['totalbill'])
        for j in cd:
            total+=int(cd[j]['totalbill'])
            item+=int(cd[j]['quantity'])
        print("\nTotal sales made: ",total)
        print("Total quantity sold: ",item)
        
    else:
        print("\nYou have Made Changes Successfully!!\nData Saved.")
        loop=0    
    
        
```
